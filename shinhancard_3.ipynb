```python
from sklearn import preprocessing

def encode_features(data_df):
    features = ["gb2"]
    for feature in features:
        label_encoder = preprocessing.LabelEncoder()
        label_encoder = label_encoder.fit(native_df[feature])
        data_df[feature] = label_encoder.transform(data_df[feature])
        
    return data_df

updated_native_df = encode_features(up_native_df)
native_df.head(3)
```

```python
print("가맹점 업종 소분류값 분포: \n", updated_native_df["gb2"].value_counts())
```

- 0 = 골프
- 1 = 관광쇼핑
- 2 = 레저프소츠
- 3 = 숙박
- 4 = 외식
- 5 = 유흥
- 6 = 종합쇼핑
- 7 = 체험
- 8 = 패션쇼핑
